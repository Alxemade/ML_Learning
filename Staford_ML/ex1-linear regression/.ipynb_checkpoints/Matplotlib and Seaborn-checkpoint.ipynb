# （1）Matplotlib简单介绍

```python
import pandas as pd
forest_fires = pd.read_csv('forestfires.csv')  # 读取csv文件 https://archive.ics.uci.edu/ml/datasets/Forest+Fires
print(forest_fires.head(10))  # 这张表记录了公园的火灾情况，x和y代表的是坐标位置，area是烧毁的面积
```

```python
import matplotlib.pyplot as plt
```

## 一般做图的步骤： 初始化绘图数据，作图，显示

##  1 散点图

```python
# 使用列表数据作为坐标轴
import matplotlib.pyplot as plt

weight = [600,135,200,233,500]

height = [90,78,45,34,21]

plt.scatter(height,weight)

plt.show()
```

```python
# 使用Series作为坐标轴

#以风速数据（wind）作为x轴，area作为y轴，做出散点图

plt.scatter(forest_fires['wind'],forest_fires['area'])  # 做出散点图

plt.title('wind speed vs fire area')  # 做出标题

plt.xlabel('wind speed when fire started')  # 做出x轴

plt.ylabel('Area consumd by fire')

plt.show()
```

## 2 折线图

```python
# 使用列表数据作为坐标轴

age = [5, 10, 15, 20, 25, 30]

height = [25, 45, 65, 75, 75, 75]

plt.plot(age, height)   # 直接使用plot画出的就是折线图

plt.title('Age vs Height')

plt.xlabel('age')

plt.ylabel('Height')

plt.show()
```

## 3 条形图

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
forest_fires = pd.read_csv('forestfires.csv')  # 读取csv文件 https://archive.ics.uci.edu/ml/datasets/Forest+Fires
area_by_month = forest_fires.pivot_table(index='month',values='area',aggfunc=np.sum)  # 输出一个透视图，按照月份进行求和

print(area_by_month)
```

```python
plt.bar(np.arange(len(area_by_month)), area_by_month['area'])   # 第一个参数是取长度，第二个参数是取面积

plt.title('Month vs Area')

plt.xlabel('month')

plt.ylabel('area')

plt.show()
```

```python
plt.barh(np.arange(len(area_by_month)), area_by_month['area'])   # 这里注意一下，这里的barh和上面的bar的区别就是他的x，y轴是调换位置的

plt.title('Month vs Area')

plt.xlabel('month')

plt.ylabel('area')

plt.show()
```

## 使用不同主题，通常使用以下几种主题：

## fivethirtyeight，ggplot，dark_background，bmh

```python
# 使用两种主题做一下对比

plt.style.use('fivethirtyeight')

plt.plot(forest_fires['rain'], forest_fires['area'])

plt.show()
```

```python
plt.style.use('ggplot')

plt.plot(forest_fires['rain'], forest_fires['area'])

plt.show()
```

# (2)Matplotlib高级应用


```python
import pandas as pd
recent_grade = pd.read_csv('recent-grads.csv')
print(recent_grade.head(5))
```

## 1 直方图

```python
# 制作工资收入中位数（median列）直方图
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


recent_grade.hist('Median')  # 默认的分成10份(bins=10),而且没有网格线,按照'Median'这个属性进行划分

recent_grade.hist('Median',bins=20,grid=False)  # 可以分成20份，而且消除网格线

columns  = ['Median','Sample_size']

recent_grade.hist(column=columns,layout=(1,2),grid=False)  # 可以一次性取出二个特性进行显示
plt.show()
```

## 2 箱型图


### 箱型图是基于五数概括法（最小值，第一个四分位数，第一个四分位数（中位数），第三个四分位数，最大值）的数据

### 的一个图形汇总，还需要用到四分位数间距IQR = 第三个四分位数 - 第一个四分位数。详情请google

### 制作箱型图使用的是pandas.DataFrame.boxplot()方法



```python
import matplotlib.pyplot as plt

# 选择两列数据
sample_size = recent_grade[['Sample_size', 'Major_category']]  # 先从recent_grade选二列数据

sample_size.boxplot(by='Major_category')



# 将X轴的坐标文字旋转90度，垂直显示

plt.xticks(rotation=90)

plt.show()
```

## 3 多图合并

### 要想找出多个变量之间的关联，就要把多个变量的变化都做在同一个图上进行比较

```python
# 将两个散点图放在一起（按颜色区分），观察是否有关联

import matplotlib.pyplot as plt

plt.scatter(recent_grade['Unemployment_rate'], recent_grade['Median'], color='red')

plt.scatter(recent_grade['ShareWomen'], recent_grade['Median'], color='blue')

plt.show()
```

# (3)Seaborn入门

### eaborn其实是在matplotlib的基础上进行了更高级的API封装，从而使得作图更加容易，在大多数

### 情况下使用seaborn就能做出很具有吸引力的图，而使用matplotlib就能制作具有更多特色的图。

### 应该把Seaborn视为matplotlib的补充，而不是替代物

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

recent_grade = pd.read_csv('recent-grads.csv')
sns.distplot(recent_grade['Median'])  # seaborn和matplot区别之一就是他默认存在一个KDE
```

## 那么Pandas与Seaborn之间有什么区别呢？

### 其实两者都是使用了matplotlib来作图，但是有非常不同的设计差异

#### 在只需要简单地作图时直接用Pandas，但要想做出更加吸引人，更丰富的图就可以使用Seaborn
#### Pandas的作图函数并没有太多的参数来调整图形，所以你必须要深入了解matplotlib
#### Seaborn的作图函数中提供了大量的参数来调整图形，所以并不需要太深入了解matplotlib

```python
# 对上图进行更多的配置

sns.set_style('dark')                # 该图使用黑色为背景色

sns.distplot(recent_grade['Median'], kde=False) # 不显示密度曲线

sns.axisgrid('Median') # 设置X轴和Y轴的坐标含义

```
